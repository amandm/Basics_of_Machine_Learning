```python
# custom gradient decent
def gradient_decent(x,y,lr = 0.05,epoch=2,batch_size = 1):
    m,b = 0.9,0.1
    weights = []
    mse = []
    n = len(x)
    for _ in range(epoch):
        f = y - (m*x +b)  # f is vector whose length is all training examples
#         print(f)
#         m = m - lr*(-2*x.dot(f).sum()/n)
        m = m - lr*(-2*x.dot(f)/n)
        b = b - lr*(-2*f.sum()/n)
        weights.append((m,b))
        mse.append(mean_squared_error(y, (m*x + b)))
    return m,b,weights,mse
```

```python
# Stocastic gradient decent randomly select samples of length batch size to update the parameters
def sgd(x,y,lr = 0.05,epoch=2,batch_size = 10):
    m,b = 0.9,0.1
    weights = []
    mse = []
    n = len(x)
    for _ in range(epoch):
        batch_index = np.random.randint(0,n,batch_size) 
        batch_x = np.take(x,batch_index)
        batch_y = np.take(y,batch_index)
        f = batch_y - (m*batch_x +b)  # f is vector whose length is all training examples
#         print(f)
#         m = m - lr*(-2*x.dot(f).sum()/n)
        m = m - lr*(-2*batch_x.dot(f)/n)
        b = b - lr*(-2*f.sum()/n)
        weights.append((m,b))
        mse.append(mean_squared_error(y, (m*x + b)))    # remember that mse is always calculated for whole dataset
    return m,b,weights,mse
```

```python
# For gradient decent 
gd_m,gd_b,gd_weights,gd_mse = gradient_decent(x,y,epoch=500)
```
