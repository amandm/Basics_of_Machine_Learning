```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
print(tf.__version__)
```

```python
x = np.array([-1.0,0,1,2,3,4,5],dtype = float)
# y = 7*x - 10
y = np.array([-17., -10.,  -3.,   4.,  11.,  18.,  25.])
```

```python
model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# https://towardsdatascience.com/gradient-descent-from-scratch-e8b75fa986cc
# Goode website decribing gradient decent from scratch
```

```python
model.fit(x, y, epochs=500)
```

```python
model.predict([10]) #should be around 19
```
